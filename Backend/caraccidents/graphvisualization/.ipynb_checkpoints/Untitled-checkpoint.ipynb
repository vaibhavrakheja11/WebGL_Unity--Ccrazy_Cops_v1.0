```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
data=pd.read_csv('../NCDB_1999_to_2014.csv')
```

```python
data['C_WDAY']=data['C_WDAY'].as_type('str')
```

```python
data=data[(data['C_YEAR'].astype(int)>=1999) & (data['C_YEAR'].astype(int)<=2014) & (data['C_MNTH'].astype(str)>=str('01')) & (data['C_MNTH'].astype(str)<=str('02'))]
```

```python
def age_replace(x):
    if x == 'NN':
        return -30
    elif x == 'UU':
        return -30
    elif x == 'XX':
        return -30
    else: 
        return x
```

```python
data['P_AGE'] = data['P_AGE'].apply(lambda x: age_replace(x))
```

```python

data['age_bracket'] = data['P_AGE'].apply(lambda x: (int(x)//10)+1)
```

```python
xz12=data.groupby(['C_HOUR']).size().to_frame().reset_index()
xz122=xz.max()
xz123=xz.min()
(xz122[0])
```

```python
data_green=""
data_red=""
data_yellow=""
```

```python
max_value=xz12.max()[0]
min_value=xz12.min()[0]
middle_value1=int((max_value-min_value)/3)
middle_value2=2*int((max_value-min_value)/3)
```

```python
result_list_temp=xz12.values.tolist()
result_list=[]
for i in range(len(result_list_temp)):
    temp={}
    temp["category"]= ""
    temp["start"]= "2019-01-10 "+result_list_temp[i][0]+":00"
    temp["end"]= "2019-01-10 "+result_list_temp[i][0]+":59"
    if(result_list_temp[i][1]<=middle_value1):
        temp['icon']=data_green
        temp["text"]= "Accidents are less than 30% of the maximum"
    elif(result_list_temp[i][1]<=middle_value2 and result_list_temp[i][1]>middle_value1):
        temp['icon']=data_yellow
        temp["text"]= "Accidents are less than 60% of the maximum"
    else:
        temp['icon']=data_red
        temp['text']="Too many accidents"
    result_list.append(temp)
result_list
```

```python
def veh_grp(x):
    if x == 'UU':
        return -1
    elif x == 'XX':
        return -1
    elif int(x) > 30:
        return 6
    elif int(x) < 30:
        return int(int(x)//5)
    else:
        return -1
```

```python
data['vehicle_group'] = data['C_VEHS'].apply(lambda x: veh_grp(x))
data['vehicle_group'].astype('int64')
```
